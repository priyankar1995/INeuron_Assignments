### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

in a for loop 

range(10) creates a number sequence from 0 to 9 

range(0,10) creates a number sequence from 0 to 9

range(0,10,1) creates a number sequence from 0 to 9 with a interval of 1


```python
# as range function directly indicates the memory location, we have to convert it into a list

print("range(10) generates:",list(range(10)))
print("range(0,10) generates:",list(range(0,10)))
print("range(0,10,1) generates:",list(range(0,10,1)))
```

### 12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.
